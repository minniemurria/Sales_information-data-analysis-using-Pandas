# Data Grouped Representation

```python
import pandas as pd
```

```python
#grouping
df.groupby("Country")
```

```python
df.groupby("Country").min()
```

```python
#group according to country as well as region
df.groupby(["Country","Region"]).sum()
```

```python
#to print only total column sum
df.groupby(["Country","Region"])["Total"].sum()
```

```python
df.head()
```

```python
#grouping by year
df.groupby("Year")["Quantity"].sum()
```

```python
df.groupby(["Country","Year"])["Quantity"].sum()
```

```python
#second way to group by year
df.groupby(df["Date of Purchase"].dt.year)["Quantity"].sum()
```

```python
#passing sum, min, max together
df.groupby(["Country","Year"])["Quantity"].agg(["sum","min","max"])
```

```python
df.groupby(df["Date of Purchase"].dt.year)["Quantity"].agg("sum")
```

```python
df.groupby(df["Date of Purchase"].dt.year)["Quantity"].agg(["sum","min"])
```

# Adding New Data-Merge,Join

```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\WIN7\Desktop\\pandas_course\\data\\sales_information.csv")
```

```python
df.head()
```

```python
#merge, join - New df - Different - overlapping
#concatenate, append - New rows/columns
```

```python
#creating a new dataframe
new_df = pd.DataFrame()
new_df["Employee Name"] = df["Sales Person"].unique()
new_df
```

```python
import random
exp = [1,2,4]
random.shuffle(exp) #shuffling the list numbers

#adding new columns to new_df
new_df["Experience In Company"] = exp
new_df["Total Experience"] = new_df["Experience In Company"] + random.randint(1,10)
new_df
```

```python
#merging the new_df with df
df.merge(new_df, left_on="Sales Person",right_on="Employee Name")
```

```python
df.merge(new_df, left_on="Sales Person",right_on="Employee Name").head()
```

```python
#dropping a column
df.merge(new_df, left_on="Sales Person",right_on="Employee Name").drop("Employee Name",axis=1)
```

```python
df.join(new_df)
```

```python
#setting index of df to Sales Person
#setting index of new_df to Employee Name
df.set_index("Sales Person", inplace=True)
new_df.set_index("Employee Name", inplace=True)
```

```python
df.head()
```

```python
new_df.head()
```

```python
df.join(new_df)
```

```python
df.reset_index()
```

# Adding New Data- Append/Concat

```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\WIN7\Desktop\\pandas_course\\data\\sales_information.csv")
```

```python
df.head()
```

```python
#merge, join - New df - Different - overlapping
#concatenate, append - New rows/columns
#>>>concatenate definition: To put things together as a conncted series:
```

```python
#Creating new dataframes
new_df = df.copy()
new_df1 = df.copy()
new_df2 = df.copy()
```

```python
#concatenate
pd.concat([new_df, new_df1, new_df2])
```

```python
pd.concat([new_df, new_df1, new_df2], ignore_index=True)
```

```python
#ading axis will make it concatenate along the columns
pd.concat([new_df, new_df1, new_df2], ignore_index=True, axis=1)
```

```python
pd.concat([new_df, new_df1, new_df2], axis=1)
```

```python
pd.concat([new_df, new_df1, new_df2], axis=0)#by default axis=0, it concatenates along the rows
```

```python
pd.concat([new_df, new_df1, new_df2], axis=0).head()
```

```python
#appending
df.append(new_df)
```

```python
df.append(new_df, axis=1)#axis doesnot work in append
```

```python
df.append([new_df, new_df1, new_df2])
```
